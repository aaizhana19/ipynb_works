## According to the statistics of 2016, 88 language centers operate in Kazakhstan. And in 2020 year only 30 have Kazakh language courses.

```python
import numpy as np  
import pandas as pd 
%matplotlib inline 
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import csv
```

```python
dataset = pd.read_excel('C:\\Users\\Айдана\\Downloads\\information about lang.centers.xlsx')
```

```python
dataset.head(20)
```

## in this chart you can see the price growth

```python
ax = dataset['Price'][:20].plot(figsize=(24,10))
ax.set_xlabel('centers number')
ax.set_ylabel('Price')
ax.set_title('Price Growth')
```

## Some statistics of dataset

```python
dataset.describe()
```

## In this stripplot you can see prices according to the center

```python
fig, ax = plt.subplots(figsize=(24,10), dpi= 80)    
sns.stripplot(dataset.centers, dataset.Price, jitter=0.25, size=30, ax=ax, linewidth=.5)


plt.title('The centers in Price', color="g",fontsize=18, fontstyle="italic", fontweight="bold", linespacing=10)
plt.show()
```

## Here the price depends on the experience of the language center on the market

```python
plt.figure(figsize=(24,10))
sns.regplot(x="experienceyear", y="Price", data=dataset,color="g",scatter_kws={"s": 80},
                 robust=True, ci=None)
```

# In one group can study from 3 to 10 people


```python
fig = plt.figure(figsize=(24,10))
plt.hist("maxpeople",data=dataset,color="mediumseagreen")
plt.style.use('ggplot')
plt.title('Max People in 1 group',alpha=0.5, color="g",fontsize=18, fontstyle="italic", fontweight="bold", linespacing=10)
plt.grid(True)
```

## The average monthly price for a Kazakh language course is 14,285 KZT Maximum amount 52,000KZT and minimum 2,500KZT

```python
fig = plt.figure(figsize=(24,10))
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
plt.bar("centers","Price",data=dataset,color="mediumvioletred")
plt.title('STD Price',alpha=0.5, color="black",fontsize=18, fontstyle="italic", fontweight="bold", linespacing=10)
plt.grid(True) 
```
