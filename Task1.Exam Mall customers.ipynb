```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import random
```

```python
data = pd.read_csv("C:\\Users\\Айдана\\Downloads\\mall-customers\\Mall_Customers.csv")
data.info()
```

```python
data.head()
```

```python
gender_group=data.groupby("Genre")["CustomerID"].sum()
```

```python
pie, ax = plt.subplots(figsize=[10,6])
labels = gender_group.keys()
plt.pie(x=gender_group,autopct="%.1f%%",explode=[0.05]*2, labels=labels, pctdistance=0.5)
plt.title("Fig1. All customers by Gender", fontsize=14);
```

```python
plt.figure(figsize=(10,6))
plt.grid(True)
plt.title('Fig2.Annual Income(k$)',fontsize=14)
plt.xlabel('Income')
plt.ylabel('Count')
sns.distplot(data['Annual Income (k$)'],color="#820556")
```

```python
import numpy as np
color_rectangle = np.random.rand(53, 3)    
plt.figure(figsize=(15,6))
plt.xlabel('Ages')
plt.ylabel('Count')
plt.title('Fig3. Age distribution')
plt.grid(True)
plt.bar(data.Age,data.CustomerID,color=color_rectangle)
```

```python
plt.figure(figsize=(10,6), dpi= 80)
sns.heatmap(data.corr(), xticklabels=data.corr().columns, yticklabels=data.corr().columns, cmap='viridis', center=0, annot=True)

plt.title('Fig4. Correlogram of the Mall Customers dataset', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,6), dpi= 60)    
sns.stripplot(data["Genre"], data['Spending Score (1-100)'], jitter=0.25, size=10, ax=ax, linewidth=.5)
plt.title('Fig5. Spending Score by gender classification', fontsize=16)
plt.show()
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(15,6))
plt.title("Fig6. Annual Income",fontsize=14)
sns.scatterplot(data=data, x="Annual Income (k$)", y="CustomerID", hue="Genre",palette="mako",legend="full",s=100)
```

```python
sns.set(style="darkgrid")
sns.pairplot(data,palette="viridis")
plt.title('Fig7. Pairplot for Mall Customers dataset', fontsize = 20)
plt.show()
```
