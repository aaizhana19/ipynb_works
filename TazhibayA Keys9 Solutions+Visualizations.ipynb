```python
DataFifa = pd.read_csv('C:\\Users\\Айдана\\Downloads\\data.csv')
DataFifa.head(5)
```

```python
DataFifa.drop(['Unnamed: 0','Photo','Flag','Club Logo','Jersey Number'],inplace=True,axis=1)
DataFifa.head(2)
```

```python
def convert(val):
    lookup = {'K': 1_000, 'M': 1_0000_00, '0':1}
    if type(val) != float:
        unit = val[-1]
        if unit in lookup:
            number = pd.to_numeric(val[:-1])
        return lookup[unit] * number
    return val
```

```python
DataFifa['Value'] = DataFifa['Value'].apply(convert)
```

```python
def rating(DataFifa):
    return int(round((DataFifa[['Potential', 'Overall']].mean()).mean()))

DataFifa['Rating'] = DataFifa.apply(rating, axis =1)
```

### 1)Отдельных игроков, которым платят значительно больше/меньше за их качество/рейтинг

```python
TOP_20 = DataFifa.nlargest(20, ['Wage','Rating']).set_index('ID')
TOP_20[['Name', 'Wage']].rename(columns={'Name': 'Players'})
```

```python
plt.figure(figsize=(12,8))
labels = TOP_20.index
ax = sns.regplot(x = 'Rating', y = 'Wage' , data = TOP_20,
            line_kws={"color":"r","alpha":1,"lw":2.5})
plt.title('The Top 20 Players who have the most high Wage(By Rating)',fontsize=14)

for i, label in enumerate(labels):
    ax.annotate(label, (TOP_20['Rating'][i],TOP_20['Wage'][i]))
```

#### 2)Отдельных клубы, которые платят значительно больше/меньше за качество/рейтинг их игроков

```python
max_wage_clubs_by_overall = DataFifa.groupby("Club", as_index=False)["Wage","Rating"].max()
max_wage_clubs_by_overall.sort_values(by="Wage",inplace=True,ascending=False)
TOP_20_C = max_wage_clubs_by_overall.iloc[0:20,:]
TOP_20_C
```

```python
import numpy as np
import matplotlib.lines as mlines
import matplotlib.transforms as mtransforms

fig, ax = plt.subplots(figsize=(15,8))
sns.scatterplot(data=TOP_20_C, x="Rating", y="Wage", palette='hsv',hue='Club',s=1000,alpha=0.5,edgecolors='black',linewidth=10)
line = mlines.Line2D([0, 1], [0, 1], color='purple')
transform = ax.transAxes
line.set_transform(transform)
ax.add_line(line)
plt.title('TOP 20 Clubs that pay significantly more for the Rating of their players',fontsize=14)
```

```python
min_wage_clubs_by_overall = DataFifa.groupby("Club", as_index=False)["Wage","Rating"].min()
min_wage_clubs_by_overall.sort_values(by="Wage",inplace=True,ascending=False)
Bottom_20_C = min_wage_clubs_by_overall.iloc[0:20,:]
Bottom_20_C
```

```python
fig, ax = plt.subplots(figsize=(15,8))
sns.scatterplot(data=Bottom_20_C, x="Rating", y="Wage", palette='hsv',hue='Club',s=1000,alpha=0.5,edgecolors='black',linewidth=10)
line = mlines.Line2D([0, 1], [0, 1], color='purple')
transform = ax.transAxes
line.set_transform(transform)
ax.add_line(line)
plt.title('TOP 20 Clubs that pay significantly Less for the Rating of their players',fontsize=14)
```

#### Задание 2 – попытайтесь построить гипотезу, объясняющую каким образом формируется рыночная стоимость игроков и проверьте её

##### Гипотеза Рыночная стоимость формируется исходя из индивидуального Рейтинга, возраста, Потенциала и других показателей отдельного игрока.

```python
DataFifa['Release Clause'] = DataFifa['Release Clause'].str.replace('€','').str.replace('M',' 1000000').str.replace('K',' 1000')
DataFifa['Release Clause'] = DataFifa['Release Clause'].str.split(' ', expand=True)[0].astype(float) * DataFifa['Release Clause'].str.split(' ', expand=True)[1].astype(float)
DataFifa['Release Clause'] = DataFifa['Release Clause'].fillna(0).astype(np.float32)
```

```python
DataFifa.corr()[['Release Clause']].sort_values(by='Release Clause', ascending=False)
```

```python
plt.figure(figsize=(8, 12))
heatmap = sns.heatmap(DataFifa.corr()[['Release Clause']].sort_values(by='Release Clause', ascending=False), vmin=-1, vmax=1, annot=True, cmap='BrBG')

heatmap.set_title('Features Correlating with Release Clause', fontdict={'fontsize':18}, pad=16);
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=DataFifa, x="Rating", y="Release Clause", hue="Body Type", size="Reactions")
plt.title('The relation between Rating and Release Clause',fontsize=14)
```

### Вывод: Рейтинг игроков имеет значение для Рыночной стоимости. Чем выше Рейтинг тем дороже Рыночная стоимость игрока.

```python
sns.lmplot(x="Age", y="Release Clause",data=DataFifa,markers="v",order=2, ci=None, scatter_kws={"color": "purple"},line_kws={"linewidth":3,"color":"red"},aspect=2);
plt.title("Age vs Release Clause",fontsize=14)
```
