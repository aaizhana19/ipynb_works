```python
class Instrument:
    counter = 0
    
    def __init__(self, name):
        self.name = name
        self.id = Instrument.counter
        Instrument.counter += 1
        
    def __str__(self):
        return self.name

class Artist:
    counter = 0
    
    def __init__(self, name, instruments):
        self.name = name
        self.instruments = instruments
        self.id = Artist.counter
        Artist.counter += 1
    
    def add_instrument(self, instrument):
        if instrument not in self.instruments:
            self.instruments.append(instrument)
    
    def remove_instrument(self, instrument):
        if instrument in self.instruments:
            self.instruments.remove(instrument)
            
    def _str_(self):
        return self.name
    
    
class Group:
    def __init__(self, name, artists=[]):
        self.name = name
        self.artists = artists
    
    def add_artist(self, artist):
        if artist not in self.artists:
            self.artists.append(artist)
        
    def remove_artist(self, artist):
        if artist in self.artists:
            self.artists.remove(artist)
    
    def count_instrument(self):
        instruments = dict()
        
        for artist in self.artists:
            for instrument in artist.instruments:
                instruments.setdefault(instrument.name, 0)
                instruments[instrument.name] += 1
        
        for key in sorted(instruments):
            print(key, '-', instruments[key], end=', ')
        
    def __str__(self):
        return self.name
    
    
```

```python
instrument1 = Instrument('Timpani')
instrument2 = Instrument('Xylophone')

instrument3 = Instrument('Cello')
instrument4 = Instrument('Bass')

```

```python
artist1 = Artist('Айжан', [instrument1, instrument2])
artist2 = Artist('Максат', [instrument3, instrument4])
artist3 = Artist('Робия', [instrument1, instrument4])

```

```python
group1 = Group('ШИМ', [artist1, artist2])
group1.add_artist(artist3)
group1.remove_artist(artist1)

```

```python
artist1.add_instrument(instrument3)
artist1.add_instrument(instrument4)

group1.count_instrument()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self, list_name):
        self.name = list_name
        self.start = None
        self.head = None
        self.length = 0
        self.end = None

    def append(self, data):
        if self.head is None:
            self.start = data
            self.head = Node(data)
            self.length += 1
        else:
            self.end = data
            new_node = Node(data)
            new_node.next = self.head
            self.head = new_node
            self.length += 1

    def append_list(self, data):
        for item in data:
            self.append(item)

    def __getitem__(self, item):
        temp = self.head
        for i in range(item):
            temp = temp.next
        return temp.data

    def first(self):
        return self.start

    def last(self):
        return self.end

    def __len__(self):
        return self.length
    
    
    def split_list_by_type(self):
        types = dict()
        temp = self.head
        
        for i in range(self.length):
            types.setdefault(type(temp.data), LinkedList(type(temp.data)))
            types[type(temp.data)].append(temp.data)
            temp = temp.next
        
        return types

    def __str__(self):
        return ("{s.name}: has {s.length} elements; \nFirst element - {s.start}; Last element {s.end}".format(s=self))


```

```python
string = "mississmithishissister"
words = ["miss", "missis", "smith", "this", "is", "his", "sister", "hiss", "mith"]


def find_combination(string, words, combination=[]):
    if string == "".join(combination):
        return combination

    for word in words:
        if string == "".join(combination):
            return combination
        if not string.startswith("".join(combination) + word):
            continue

        combination.append(word)
        res = find_combination(string, words, combination)
        if res:
            return res
        else:
            combination.pop()
    return False


print(find_combination(string, words))
```
